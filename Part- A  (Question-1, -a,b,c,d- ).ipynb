```python
import pandas as pd
```

```python
gameplay = pd.read_csv('User Gameplay data.csv')
withdraw = pd.read_csv('Withdrawal Data.csv')
deposit = pd.read_csv('Deposit Data.csv')
```

```python
gameplay = gameplay.rename(columns={'User ID': 'User Id'})
```

```python
gameplay
```

```python
gameplay['Datetime'] = pd.to_datetime(gameplay['Datetime'], format= '%d-%m-%Y %H:%M')
```

```python
gameplay.set_index('Datetime', inplace=True)
```

```python
gameplay_data= gameplay['2022-10-26 12:00':'2022-10-26 23:59'].groupby('User Id').size().reset_index(name = 'no. of games played')
```

```python
gameplay_data
```

```python
withdraw
```

```python
withdraw= withdraw.rename(columns={'Amount':'Withdraw Amount'})
```

```python
withdraw['Datetime'] = pd.to_datetime(withdraw['Datetime'], format= '%d-%m-%Y %H:%M')
```

```python
withdraw.set_index('Datetime', inplace=True)
```

```python
withdraw
```

```python
filtered_data = withdraw['2022-10-26 12:00':'2022-10-26 23:59']
```

```python
withdrawn= filtered_data.groupby('User Id').agg({'Withdraw Amount': 'sum'}).reset_index()
```

```python
num_times_withdrawn = filtered_data.groupby('User Id').size().reset_index(name='no. of times withdrawn')
```

```python
num_times_withdrawn.head()
```

```python
withdraw_data = pd.merge(withdrawn,num_times_withdrawn, on='User Id')
```

```python
withdraw_data
```

```python
deposit['Datetime']= pd.to_datetime(deposit['Datetime'], format= '%d-%m-%Y %H:%M')
```

```python
deposit.set_index('Datetime', inplace=True)
```

```python
deposit= deposit.rename(columns={'Amount': 'Deposit Amount'})
```

```python
deposit
```

```python
filter_data= deposit['2022-10-26 12:00':'2022-10-26 23:59']
```

```python
total_deposit_amount = filter_data.groupby('User Id').agg({'Deposit Amount':'sum'}).reset_index()
```

```python
num_times_deposit = filter_data.groupby('User Id').size().reset_index(name='no. of times deposit money')
```

```python
num_times_deposit
```

```python
deposit_data = pd.merge(total_deposit_amount, num_times_deposit, on='User Id')
```

```python
deposit_data
```

```python
merged = pd.merge(gameplay_data, deposit_data, on='User Id', how='outer')
```

```python
gaming_data = pd.merge(merged, withdraw_data, on='User Id', how='outer').fillna(0)
```

```python
gaming_data
```

- S1 from 12am to 12pm 
- S2 from 12pm to 12am

# 1. Find Playerwise Loyalty points earned by Players in the following slots:-

# 2nd October Slot S1

```python
gaming_data['Loyalty Points'] = (0.01 * gaming_data['Deposit Amount']) + (0.005 * gaming_data['Withdraw Amount']) + (0.001 * (max(gaming_data['no. of times deposit money']- gaming_data['no. of times withdrawn']) or 0)) + (0.2 * gaming_data['no. of games played'])
```
